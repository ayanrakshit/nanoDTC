## **Manual** climate probability model

This dataset has one record per station per month. Different stations have different lengths of records: some start at 1853, the most recent starts in 1978. Source: the UK Met Office. See the accompanying file [climate.ipynb](climate.ipynb) for scraping code.

```python
# url = 'https://www.cl.cam.ac.uk/teaching/2122/DataSci/data/climate.csv'
url = 'climate.csv'
climate = pandas.read_csv(url)

# Define a timestamp, measured in years
climate['t'] = climate.yyyy + (climate.mm-1)/12

# Restrict attention to recent records from Cambridge
climate = climate.loc[(climate.station=='Cambridge') & (climate.yyyy>=1985)]

# How long do the records go back, for each station?
climate.groupby('station').apply(lambda df: min(df.yyyy)).sort_values()
```

Here is an abitraty probability model for temperature, first in random variable notation and below in code:
$$ \textit{Temp}_i = \alpha \sin\bigl(2\pi(t+\phi)\bigr) + c + \gamma t + \operatorname{Normal}(0,\sigma^2) $$

We can choose any plausible values for the unknown parameters.
Plug in different parameters and see if you can get a better fit.

```python
def rtemp(t, α=9, φ=-0.25, c=10, γ=0, σ=2):
    pred = α*np.sin(2*π*(t+φ)) + c + γ*t
    return np.random.normal(loc=pred, scale=σ)
```

```python
fig,(ax,ax2) = plt.subplots(2,1, figsize=(15,4), sharex=True, sharey=True)

# Plot the (noiseless) predicted sinusoid
newt = np.linspace(1985, 2025, 1000)
ax.plot(newt, rtemp(newt, σ=0), color='0.7')
ax2.plot(newt, rtemp(newt, σ=0), color='0.7')

# Scatter-plot the actual data
ax.scatter(climate.t, climate.temp, color='black', marker='+')
ax.set_title('dataset')

# Scatter-plot some simulated data
for _ in range(60):
    ax2.scatter(climate.t, rtemp(climate.t), color='cornflowerblue', marker='o', s=3, alpha=0.1)
ax2.set_title('simulation')
    
plt.show()
```

## **Fitted** climate probability model

```python
# Define two extra covariates
climate['t'] = climate.yyyy + (climate.mm-1)/12    # timestamp in years
climate['temp'] = (climate.tmin + climate.tmax)/2  # monthly avg temp
```

## A periodic pattern


```python
t, temp = climate['t'], climate['temp']

# Set up the variables vectors
X = np.column_stack([np.sin(2*π*t), np.cos(2*π*t)])
model0 = sklearn.linear_model.LinearRegression()
model0.fit(X, temp)
α,(β1,β2) = model0.intercept_, model0.coef_
α,(β1,β2)
```

```python
newt = np.linspace(1985, 2024, 1000)
pred = model0.predict(np.column_stack([np.sin(2*π*newt), np.cos(2*π*newt)]))

fig,ax = plt.subplots(figsize=(15,1.7))
ax.plot(newt, pred, color='0.7', zorder=0)
plt.scatter(t, temp, s=15, alpha=0.7)
plt.xlim(1984, 2025)
plt.title(f'$\\alpha={α:.3}, \\quad\\beta_1={β1:.3}, \\quad\\beta_2={β2:.3}$')
plt.ylabel('mean temp')
plt.show()
```

## A Secular trend

```python
t, temp = climate['t'], climate['temp']

# Add an extra variable dependence
X_A = np.column_stack([np.sin(2*π*t), np.cos(2*π*t), t])
modelA = sklearn.linear_model.LinearRegression()
modelA.fit(X_A, temp)
α,(β1,β2,γ) = modelA.intercept_, modelA.coef_
```

```python
newt = np.linspace(1985, 2024, 1000)
pred = modelA.predict(np.column_stack([np.sin(2*π*newt), np.cos(2*π*newt), newt]))

fig,ax = plt.subplots(figsize=(15,1.7))
ax.plot(newt, pred, color='0.7', zorder=0)
plt.scatter(t, temp, s=15, alpha=0.7)
plt.xlim(1984, 2025)
plt.title(f'α={α:.3}, γ={γ:.3}')
plt.ylabel('mean temp')
plt.show()
```

## Understanding the residuals plot

```python
# Model 0
ε = temp - model0.predict(X)

fig,ax = plt.subplots(figsize=(15,1.7))
ax.axhline(0, color='0.8', zorder=-1)
ax.scatter(t[ε>0], ε[ε>0], color='red', s=15, alpha=0.7)
ax.scatter(t[ε<=0], ε[ε<=0], color='blue', s=15, alpha=0.7)

plt.show()

# Model A
ε = temp - modelA.predict(X_A)

fig,ax = plt.subplots(figsize=(15,1.7))
ax.axhline(0, color='0.8', zorder=-1)
ax.scatter(t[ε>0], ε[ε>0], color='red', s=15, alpha=0.7)
ax.scatter(t[ε<=0], ε[ε<=0], color='blue', s=15, alpha=0.7)

plt.show()
```
